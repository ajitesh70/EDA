```python
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
df
```

```python
df.info()
```

```python
df['packed_cell_volume']=pd.to_numeric(df.packed_cell_volume,errors='coerce')  #coerce >> if there are null values it will ignore them 
```

```python
df['white_blood_cell_count']=pd.to_numeric(df.white_blood_cell_count,errors='coerce')
```

```python
df['red_blood_cell_count']=pd.to_numeric(df.red_blood_cell_count,errors='coerce')
```

```python
cat_col=[col for col in df.columns if df[col].dtype=='object']
```

```python
num_col=[col for col in df.columns if df[col].dtype!='object']
```

```python
cat_col
```

```python
num_col
```

```python
for col in cat_col:
    print(f"{col} has {df[col].unique()} value. \n")
```

```python
df['diabetes_mellitus'].replace(to_replace={'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)
```

```python
df['coronary_artery_disease'].replace(to_replace={'\tno':'no'},inplace=True)
```

```python
df['class'].replace(to_replace={'notckd':'not ckd'},inplace=True)
```

```python
df['class']=df['class'].map({'ckd':0, 'not ckd':1})
```

```python
def kde(col):
    grid=sns.FacetGrid(df,hue="class",height=6,aspect=2)
    grid.map(sns.kdeplot,col)
    grid.add_legend()
    plt.show()
```
